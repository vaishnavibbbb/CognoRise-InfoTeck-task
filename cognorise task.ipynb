```python
import pandas as pd
import matplotlib.pyplot as plt

# Example assuming the dataset structure
# Load your dataset (replace with your actual dataset loading code)
# df2 = pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')

# Example 1: Time series plot of unemployment rate from df2
# Assuming column names are 'YearMonth' and 'Unemployment_Rate'
df2 = pd.DataFrame({
    'YearMonth': ['2020-01', '2020-02', '2020-03', '2020-04', '2020-05'],
    'Unemployment_Rate': [5.2, 5.5, 6.0, 6.3, 6.1]
})

plt.figure(figsize=(12, 6))
plt.plot(df2['YearMonth'], df2['Unemployment_Rate'], marker='o', linestyle='-')
plt.title('Unemployment Rate Over Time (Up to Nov 2020)')
plt.xlabel('Year-Month')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
